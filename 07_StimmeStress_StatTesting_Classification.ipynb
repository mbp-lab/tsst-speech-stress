# Statistical testing (Corrected t-test) for the models with nested cv hyperparameter tuning

Here we are using the corrected paired t-test proposed by Nadeau and Bengio to compare the model performance against the baseline (majority vote in training split)

```python
RF_score_csvs = ["LR-nested_kfold_fold_scores.csv", "rf - nested_classification_kfold_fold_scores.csv",
                 "svm - nested_classification_kfold_fold_scores.csv", "xgb - nested_classification_kfold_fold_scores.csv"]

for csv in RF_score_csvs:
    df = pd.read_csv(csv)

    differences = df["model_score"].values - df["baseline_score"].values

    n = differences.shape[0]
    df = n - 1
    n_train = 45
    n_test  = 5    

    t_stat, p_val = mystats.compute_corrected_ttest(
        differences, df, n_train, n_test
    )
    print(csv)
    print("----------------------")
    print(f"Corrected t-value: {t_stat:.3f}\nCorrected p-value: {p_val:.5f}\n")
    print()
```
