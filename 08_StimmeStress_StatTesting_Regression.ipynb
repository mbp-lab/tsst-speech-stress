# Statistical testing (Corrected paired t-test)

Here we are using the corrected paired t-test proposed by Nadeau and Bengio to compare the model performance against the baseline (mean absolute value)

## For models used only TSST data

### Random Forest Regressor

```python
RF_score_csvs = [
    "RF_cortisol_react_loo_abs_error.csv",
    "RF_cort_20_min_loo_abs_error.csv",
    "RF_saa_react_loo_abs_error.csv",
    "RF_pana_delta_na_loo_abs_error.csv",
    "RF_pana_delta_pa_loo_abs_error.csv"
]

for csv in RF_score_csvs:
    df = pd.read_csv(csv)

    differences = df["baseline_abs_error"].values - df["model_abs_error"].values

    n = differences.shape[0]
    dfree = n - 1
    n_train = 24
    n_test = 1    

    t_stat, p_val = mystats.compute_corrected_ttest(
        differences, dfree, n_train, n_test
    )

    mean_abs_error_model = np.mean(df["model_abs_error"].values)
    mean_abs_error_baseline = np.mean(df["baseline_abs_error"].values)

    print(csv)
    print("----------------------")
    print(f"Corrected t-value: {t_stat:.3f}")
    print(f"Corrected p-value: {p_val:.3f}")
    print(f"Mean difference (abs): {np.mean(differences)}")
    print(f"Mean abs error model: {mean_abs_error_model}")
    print(f"Mean abs error baseline: {mean_abs_error_baseline}")
    print()
```
